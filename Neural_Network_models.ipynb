<h1><center>PCA</center></h1>

```python
## import needed packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## load preprocessed dataset

df = pd.read_csv('Preprocessed.csv')
df.head()
```

```python
X = df.drop('subscribed', axis=1)

y = df['subscribed']
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the features
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
PrincipalComponents = pca.fit_transform(X)
pca_df = pd.DataFrame(data = PrincipalComponents, columns = ['pc1', 'pc2'])
```

```python
pca_df.head()
```

```python
final_df = pd.concat([pca_df, y], axis = 1)
```

```python
final_df.head()
```

```python
plt.figure(dpi=150)
ax = sns.scatterplot(data=final_df, x="pc1", y="pc2", hue="subscribed")
plt.show()
```

<h1><center>MODELS</center></h1>

```python
## import packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# load data

df = pd.read_csv("PCA_Bank_marketing.csv")
df.head()
```

```python
X = df.drop('subscribed', axis=1)
y = df['subscribed']
```

```python
from sklearn.model_selection import train_test_split

#help(train_test_split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

<h1><center>CNN</center></h1>

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(2,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train,
          batch_size=32, epochs=100,
          validation_data=(X_test, y_test))
```

```python
# graph of loss function 

plt.figure(figsize=(6, 3), dpi=150)
plt.plot(hist.history['loss'])

plt.xlabel("epochs")
plt.ylabel("loss")

plt.title("loss vs epochs")
plt.show()
```

```python
# graph of loss function 

plt.figure(figsize=(6, 3), dpi=150)

plt.plot(hist.history['accuracy'])
plt.xlabel("epochs")
plt.ylabel("accuracy and accuracy")
plt.legend()
plt.title("accuracy vs epochs")
plt.show()
```

<h1><center>RNN</center></h1>

```python
#import libraries and packages:
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import SimpleRNN
from keras.layers import Dropout

#Initialize RNN:
model_RNN = Sequential()

#Adding the first RNN layer and some Dropout regularization
model_RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True, input_shape= (X_train.shape[1],1)))
model_RNN.add(Dropout(0.2))

#Adding the second RNN layer and some Dropout regularization
model_RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True))
model_RNN.add(Dropout(0.2))

#Adding the third RNN layer and some Dropout regularization
model_RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True))
model_RNN.add(Dropout(0.2))

#Adding the fourth RNN layer and some Dropout regularization
model_RNN.add(SimpleRNN(units = 50))
model_RNN.add(Dropout(0.2))

#Adding the output layer
model_RNN.add(Dense(units = 1))

#Compile the RNN
model_RNN.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

#Fitting the RNN to the Training set
history_RNN = model_RNN.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
# graph of loss function 

plt.figure(figsize=(6, 3), dpi=150)
plt.plot(history_RNN.history['loss'])

plt.xlabel("epochs")
plt.ylabel("loss")
#plt.legend()
plt.title("loss vs epochs")
plt.show()
```

```python
# graph of loss function 

plt.figure(figsize=(6, 3), dpi=150)
plt.plot(history_RNN.history['accuracy'])

plt.xlabel("epochs")
plt.ylabel("loss")
#plt.legend()
plt.title("loss vs epochs")
plt.show()
```

<h1><center>Kmeans</h1></center>

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
# load data

df = pd.read_csv("PCA_Bank_marketing.csv")
df.head()
```

```python
X = df.drop('subscribed', axis=1)
```

```python
kmeans = KMeans(2)
kmeans.fit(X)
```

```python
identified_clusters = kmeans.fit_predict(X)
identified_clusters
```

```python
df_with_clusters = df.copy()
df_with_clusters['Clusters'] = identified_clusters 
plt.scatter(df_with_clusters['pc1'],df_with_clusters['pc2'],c=df_with_clusters['Clusters'],cmap='rainbow')
```


WCSS is the sum of squared distance between each point and the centroid in a cluster. When we plot the WCSS with the K value, the plot looks like an Elbow. As the number of clusters increases, the WCSS value will start to decrease. WCSS value is largest when K = 1

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(X)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
