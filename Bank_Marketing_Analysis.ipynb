```python
## import packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## import data

df = pd.read_csv('bank_marketing_dataset.csv')
```

```python
## Count plot

def count_plot(name):
    """
    This is helps to plot the count plot
    """
    
    plt.figure(dpi=150)
    ax = sns.countplot(x=name, data=df)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
## count plot for "job"

count_plot('job')
```
